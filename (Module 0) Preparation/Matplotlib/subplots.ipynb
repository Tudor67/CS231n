```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2, 1, 1)
plt.plot(x, y_sin)
plt.title('sin')

plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('cos')

plt.show()
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.1, 100, 0.1)
y_exp = np.exp(x)
y_log = np.log(x)

plt.subplot(2, 1, 1, facecolor='orange')
plt.plot(x, y_exp, color='g', linewidth=3)
plt.title('exp')

plt.subplot(2, 1, 2, facecolor='blue')
plt.plot(x, y_log, color='white')
plt.title('log')

plt.tight_layout()
plt.show()
```
