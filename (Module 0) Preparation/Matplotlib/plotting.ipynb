```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Sine and Cosine')
plt.legend(['Sine', 'Cosine'])
plt.show()
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, 'r--', label='sin')
plt.plot(x, y_cos, 'g.', label='cos')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('sin and cos')
plt.legend()
plt.axis([-1, 13, -2, 2])
plt.show()
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import os

dir_name = 'plots'
if not os.path.isdir(dir_name):
    os.mkdir(dir_name)

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.gca().set_facecolor('y')
plt.plot(x, y_sin, label='sin',
         linewidth='4', color='g', 
         marker='o', markerfacecolor='r', markersize=5)

plt.plot(x, y_cos, label='cos',
         linewidth='7', color='b',
         marker='^', markerfacecolor='k', markersize=8)

plt.title('sin and cos')
plt.legend()
plt.savefig('./' + dir_name + '/test_fig.png')

plt.show()
```
